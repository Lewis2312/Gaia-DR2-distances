Enter the parallax and parallax uncertainty below, as well as either the prior length scale (rlen) or the Galactic longitude and latitude of the star. If the latter are specified the prior length scale is computed from the Galactic model, as done in the paper and for the GDR2 catalogue. (If both are specified, rlen takes precedence.)

The code will write various statistics to the screen and plot the posterior (black or red line) and prior (green line).
* If the prior is unimodal the posterior is plotted in black.
* If the prior is bimodal the posterior is plotted in red.
* The green line may not be visible if the posterior and prior are near identical or if the posterior is very narrow and tall compared to the prior (the latter can then be constant at near zero density over the range of the plot)
The solid vertical line is the distance estimator, and the dashed vertical lines are the lower and upper bounds of the (approximately) 68% confidence interval. 
When result_flag=1 these are the mode and HDI (highest density interval)
When result_flag=2 these are the median and ETI (equal-tailed interval)

If the resulting plot has a bad range, or even fails to plot, adjust the plotting range using rplotlo,rplothi (specified below). For the purpose of plotting (only) the posterior is normalized (numerically) over the range specified by rlo,rhi below. These must extend over the full range of the posterior! But if the posterior is *much* narrower than this range, the normalization may fail.

When result_flag=2 and modality_flag=2 (distribution is bimodal), the code uses MCMC to compute the quantiles (progress of this is written to the screen). An additional plot is produced in this case showing a histogram of the MCMC samples compared to the actual posterior, so you can judge how likely the quantiles are a suitable summary of the posterior.

If result_flag=0 the code could not compute either estimator and the reason for this will be reported. The plotting may also. See the paper for more details.
